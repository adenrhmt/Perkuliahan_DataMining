10 Artis Teratas berdasarkan Total Grammy Awards

```python
# Sort the songs by total streams in descending order
df1.sort_values("Main Grammy awards", ascending=False, inplace=True)

# Plot a bar graph to visualize the top songs based on streams
plt.figure(figsize=(20,12))
plt.bar(df1["Artist"][:10], df1["Main Grammy awards"][:10])

# Add labels and title to the graph
plt.xlabel("Artist")
plt.ylabel("Main Grammy Award")
plt.title("Top 10 Artists based on Total Grammy Awards")

# Show the plot
plt.show()
```

```python
# Create an interactive scatter plot of the top songs based on position and total streams
fig = px.scatter(df1.sort_values("Main Grammy awards", ascending=False)[:30], x='Main Grammy Nominations', y='Main Grammy awards', color='Artist', size='Main Grammy awards',
           hover_name='Artist', title='Top Artist by Total Grammy Awards')

fig.show()
```
