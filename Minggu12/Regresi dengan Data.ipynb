```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
dataset = pd.read_excel('ElectricityConsumptionAndProduction.xlsx')
x = dataset.iloc[:, 1].values
y = dataset.iloc[:, 2].values
```

```python
dataset.keys()
```

```python
dataset.shape
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train.reshape(-1, 1), y_train.reshape(-1, 1))
```

```python
y_pred = regressor.predict(x_test.reshape(-1, 1))
```

```python
plt.scatter(dataku.Consumption, dataku.Production)
plt.xlabel("Konsumsi")
plt.ylabel("Produksi")
plt.title("Grafik Konsumsi vs Produksi")
plt.show
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x_train.reshape(-1, 1), y_train.reshape(-1, 1), color='blue')
plt.plot(x_train.reshape(-1, 1), regressor.predict(x_train.reshape(-1, 1)), color='red')
plt.title('Konsumsi Terhadap Produksi (Training Set)')
plt.xlabel('Konsumsi')
plt.ylabel('Produksi')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x_test.reshape(-1, 1), y_test.reshape(-1, 1), color='blue')
plt.plot(x_test.reshape(-1, 1), regressor.predict(x_test.reshape(-1, 1)), color='red')
plt.title('Konsumsi Terhadap Produksi (Test Set)')
plt.xlabel('Konsumsi')
plt.ylabel('Produksi')
plt.show()
```
