```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
x = np.array([[2,6],[8.2,9.3],[5,6],[3,1.5],[7,3.7]])
```

```python
print(x)
```

```python
Z = linkage(x)
```

```python

print(Z)
```

```python
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('Data')
plt.ylabel('Jarak')
dendrogram(Z)
plt.show()
     
```
